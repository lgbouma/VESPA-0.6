```python
class A(object):
    def __init__(self):
        pass
    
    @property
    def __prop(self):
        return 'a'
    
class B(object):
    def __init__(self):
        pass
    
    @property
    def __prop(self):
        return 'b'
    
class C(A,B):
    def __init__(self):
        B.__init__(self)
        pass
    
    @property
    def __prop(self):
        return 'c' + super(C,self)._A__prop + self._B__prop
```
